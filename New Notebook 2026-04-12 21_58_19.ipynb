```python
data= [(1,"Phone",200,"South"),
        (2,"Laptop",800,"North"),
       (3,"Tablet",300,"East"),
       (4,"Phone",150,"South"),
       (5,"Laptop",900,"West"),
       (6,"Tablet",400,"North"),
       (7,"Phone",250,"East"),
       (8,"Laptop",1000,"South"),
       (9,"Tablet",350,"West"),
       (10,"Phone",180,"North")
       ]
columns = ["order_id","product","amount","region"]
df=spark.createDataFrame(data,columns)
df.display()
```

```python
from pyspark.sql.functions import col, sum

#Filter high-value orders
df_filtered = df.filter(col("amount") > 300)

#Aggreagate by region
df_grouped = df_filtered.groupBy("region").agg(sum("amount").alias("region wise total amount"))
display(df_grouped)
```

```python
df_grouped.write.mode("overwrite").saveAsTable("sales_summary")
```
