# **CI/CD: Continues Integration and continues Deployment or continues development**
CI:
Developers frequently push code to a shared repository (GitHub, GitLab, Azure Repos)
- Code is built
- Unit tests run
- Errors are caught early

CI = Build + Test

CD:(Deployment)
After CI passes, the code is automatically prepared for release
- Code is deployed to staging / test environment
- Production release is manual (someone clicks Deploy)

CD:(Development)
If tests pass → code is automatically deployed to production no manual approval needed

```python
data= [(1,"Phone",200,"South"),
        (2,"Laptop",800,"North"),
       (3,"Tablet",300,"East"),
       (4,"Phone",150,"South"),
       (5,"Laptop",900,"West"),
       (6,"Tablet",400,"North"),
       (7,"Phone",250,"East"),
       (8,"Laptop",1000,"South"),
       (9,"Tablet",350,"West"),
       (10,"Phone",180,"North")
       ]
columns = ["order_id","product","amount","region"]
df=spark.createDataFrame(data,columns)
df.display()
```

```python
from pyspark.sql.functions import col, sum

#Filter high-value orders
df_filtered = df.filter(col("amount") > 100)

#Aggreagate by region
df_grouped = df_filtered.groupBy("region").agg(sum("amount").alias("region_wise_total_amount"))
display(df_grouped)
```

```python
df_grouped.write.mode("overwrite").saveAsTable("sales_summary")
```
